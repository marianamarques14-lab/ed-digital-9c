```python
vamos criar um projeto para adivinhar um numero entre 1 a 10. O usuaio devera digitar qual e o numero qaue acredita ser o correto, e entao o codigo retornara qual foi o numero pensado.
```

```python
import random

# Gerar um número aleatório entre 1 e 10
numero_secreto = random.randint(1, 10)

# Pedir ao usuário para adivinhar o número
try:
    palpite_do_usuario = int(input("Adivinhe um número entre 1 e 10: "))
except ValueError:
    print("Entrada inválida. Por favor, digite um número.")
    palpite_do_usuario = None

# Verificar o palpite do usuário
if palpite_do_usuario is not None:
    print(f"Você adivinhou: {palpite_do_usuario}")
    print(f"O número pensado era: {numero_secreto}")

    if palpite_do_usuario == numero_secreto:
        print("Parabéns! Você acertou!")
    else:
        print("Que pena! Você errou.")
```
