```python
def run_quiz(questions):
    score = 0
    num_questions_to_win = 10

    print("Bem-vindo ao Quiz da Copa do Mundo!\n")
    print(f"Acerte {num_questions_to_win} perguntas para ganhar o jogo.\n")

    while True:
        if score == num_questions_to_win:
            print("\nParabéns! Você acertou 10 perguntas e ganhou o jogo!")
            break

        question_data = questions[score] # Use score as index for simplicity, or pick random from remaining
        question_text = question_data["question"]
        options = question_data["options"]
        correct_answer = question_data["answer"]

        print(f"Pergunta {score + 1}: {question_text}")
        for i, option in enumerate(options):
            print(f"{i + 1}. {option}")

        try:
            user_answer_index = int(input("Digite o número da sua resposta: ")) - 1
            if 0 <= user_answer_index < len(options):
                user_choice = options[user_answer_index]
                if user_choice == correct_answer:
                    score += 1
                    print(f"Correto! Sua pontuação atual: {score}\n")
                else:
                    print(f"Errado! A resposta correta era: {correct_answer}")
                    print("Jogo reiniciado.\n")
                    score = 0 # Reset score on wrong answer
                    random.shuffle(questions) # Reshuffle questions for a new game
            else:
                print("Opção inválida. Por favor, digite um número entre 1 e {len(options)}.\n")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.\n")

# Run the quiz
run_quiz(questions)
```
