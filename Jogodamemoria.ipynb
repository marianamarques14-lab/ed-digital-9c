## Jogo da Memória Numérica

Este jogo desafia sua memória. Você terá que memorizar uma sequência numérica que será exibida por alguns segundos. A cada acerto, a sequência ficará mais longa e o tempo de exibição aumentará. O objetivo é acertar 10 rodadas para vencer!

```python
import random
import time

def generate_sequence(length):
    """Gera uma sequência de números aleatórios com o comprimento especificado."""
    return [random.randint(0, 9) for _ in range(length)]

def display_sequence(sequence, display_time):
    """Exibe a sequência para o usuário por um tempo determinado."""
    print(f"Memorize esta sequência: {''.join(map(str, sequence))}")
    time.sleep(display_time)
    # Clear the sequence from the screen for a better challenge (optional)
    # You might need a more advanced clear screen method for actual terminal/console games
    print('\n' * 50) # A simple way to 'clear' the screen in Colab/Jupyter

def get_user_input():
    """Obtém a entrada do usuário e tenta converter para uma lista de inteiros."""
    while True:
        try:
            user_sequence_str = input("Digite a sequência que você memorizou (sem espaços): ")
            user_sequence = [int(digit) for digit in user_sequence_str]
            return user_sequence
        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números.")
```

```python
def play_memory_game():
    initial_sequence_length = 5
    initial_display_time = 5 # seconds
    max_rounds = 10

    current_sequence_length = initial_sequence_length
    current_display_time = initial_display_time
    score = 0

    print("\n----- Jogo da Memória Numérica -----")
    print("Tente memorizar a sequência de números e digitá-la corretamente.")
    print(f"Você precisa acertar {max_rounds} rodadas para vencer!")

    while score < max_rounds:
        print(f"\nRodada {score + 1}. Tamanho da sequência: {current_sequence_length}")
        print(f"Tempo de exibição: {current_display_time} segundos")

        sequence = generate_sequence(current_sequence_length)
        display_sequence(sequence, current_display_time)

        user_attempt = get_user_input()

        if user_attempt == sequence:
            print("Parabéns! Você acertou!")
            score += 1
            current_sequence_length += 1
            current_display_time += 1
        else:
            print(f"Que pena! Você errou. A sequência correta era: {''.join(map(str, sequence))}")
            print("O jogo será reiniciado.")
            current_sequence_length = initial_sequence_length
            current_display_time = initial_display_time
            score = 0 # Reset score
            print("\n----- Reiniciando o Jogo -----")

    print("\nPARABÉNS! VOCÊ GANHOU O JOGO! Acertou 10 rodadas!")

# Para iniciar o jogo, execute a função:
play_memory_game()
```
